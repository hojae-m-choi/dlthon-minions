```python
# zip  
# %%timeit
import os
path = os.path.join(DATA_DIR, 'TL_01_KAKAO.zip')
output_dir = os.path.join(DATA_DIR, "aihub_dataset")
os.system("unzip "+path+" -d "+output_dir)
```

```python
# text    
import json
unziped_dir = os.path.join(DATA_DIR, "aihub_dataset")
file_list = os.listdir(unziped_dir)
# 1000 text 
file_list = file_list[:10000]
data_list = []

count = 0
for file in file_list:
    file_path = os.path.join(unziped_dir, file)
    with open(file_path, 'r', encoding='UTF-8') as f:
        json_data = json.load(f)
        info = json_data['info'][0]
        anno = info['annotations']
        lines = anno['lines']
        if anno['speaker_type'] == '1:1':
            count += 1
            data_list.append(dict(
                id = info['id'],
                subject = anno['subject'],
                speaker_type = anno['speaker_type'],
                conversation = '\n'.join([line['norm_text'] for line in lines]),
            ))
        elif count > 1001:
            break
#         text_list.append(text)

# print(data_list)
```

```python
import pandas as pd
df = pd.DataFrame(data_list)
```
